```python
import pandas as pd
from tqdm import tqdm
```

```python
tickers = pd.read_csv("tickers.csv")
```

```python
import yfinance as yf

# List of stock symbols in the S&P 500 (replace this list with the actual symbols)

# Define the start and end dates for the data
start_date = '2020-01-01'
end_date = '2021-12-31'

# Download historical stock price data for each symbol
stock_data = {}
for symbol in tqdm(tickers["Symbol"]):
    stock = yf.Ticker(symbol)
    data = stock.history(start=start_date, end=end_date)
    stock_data[symbol] = data

# data for GEHC and CEG are not available as they are not listed at the time

# appending the result in a data frame
result_df = pd.DataFrame()
for ticker in stock_data.keys():
    temp_df = pd.DataFrame(stock_data[ticker])
    temp_df["ticker"] = ticker
    result_df = pd.concat([result_df, temp_df])

result_df.index = result_df.index.strftime('%Y%m%d')
```
